## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке). Кроме того, в исключительных случаях, при демонстрации участником на защите превосходящих ожидания уровней понимания, участия и вовлечения в рамках проекта, комиссия может своим решением повысить индивидуальную оценку для этого участника на 0,5 или 1 балл относительно общей оценки за проект (первой оценки).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(14042026)`.

Обратите, пожалуйста, внимание на то, что бОльшая часть комиссий осуществляет приём проектов в очной форме, однако могут присутствовать и 1-2 комиссии, которые будут работать дистанционно с использованием средств видеосвязи. В случае попадания команды путем случайного распределения в комиссию, работающую в очной форме, *не допускается* какое-либо изменение формата проведения для команды на онлайн-режим, равно как и любые другие перемещения команд между комиссиями (смены комиссий), для сохранения честности и непредвзятости случайного распределения.

Явка студента на защиту проекта является обязательной; вклад студента без его присутствия не оценивается и автоматически является равным нулю. Для дистанционных комиссий, дополнительным требованием является обязательное наличие включенной камеры в течение всего времени нахождения в конференции

**Важно**: демонстрация работоспособности решения на защитах является обязательной. При отсутствии у команды возможности показать, как работает агент, комиссия имеет право существенно (вплоть до 0) снизить общую оценку за проект

В современном бизнесе критически важно не только разрабатывать эффективные алгоритмы для анализа данных, но и уметь быстро адаптировать их под новые задачи. Теперь вам предстоит автоматизировать сам процесс анализа данных и решить классическую ML-задачу (задачи) с помощью ИИ-агентов.

В данном проекте вы разработаете ИИ-агента, который автономно выполняет весь пайплайн решения ML-задачи: от получения данных до сравнительной оценки работы ML-моделей.




### Постановка задачи

Вам предстоит спроектировать и реализовать ИИ-агента, который решает задачу классического машинного обучения (бинарная или многоклассовая классификация, регрессия, кластеризация) или совокупность задач на основе данных, которые он получает либо из предоставленного источника, либо из собранных вами данных (на ваш выбор)

Для начала, как и всегда, вам необходимо определиться с источником данных и бизнес-задачей. Какую боль вы хотите решить, какой процесс автоматизировать? Кто выступает вашим бизнес-заказчиком?

После этого можно переходить к сбору данных. Вы можете выбрать один или несколько способов получить данные для последующей обработки:
* Скачать любой публичный датасет из таких источников как Kaggle, UCI или Hugging Face Datasets

* Найти несколько датасетов со схожей предметной областью, которые можно объединить по общей колонке. **Важно**: соединение и дальнейшую обработку должен выполнять ИИ-агент на последующих этапах

* Вы можете воспользоваться изученными ранее инструментами для извлечения данных (bs4, requests, Selenium и другие) и собрать датасет сами. **Важно**: если вы получили несколько готовых датасетов, то соединение и дальнейшую обработку должен выполнять ИИ-агент на последующих этапах


Следующим шагом сформулируйте задачу машинного обучения и целевую переменную. Целевая переменная должна быть осмысленной с бизнес-точки зрения.

Переходим к основному этапу — реализации ИИ-агента на low-code платформе или на Python с помощью фреймворков для разработки. Ваш агент должен выполнить следующие **обязательные этапы**:

- Автоматическая загрузка данных
- Описание датасета и первичное погружение в данные (предметная область, размер и содержимое датасета, типы данных, статистические характеристики и т.д.)
- Предобработка данных. Агент должен проверить данные на пропуски, выбросы, аномалии, дисбаланс классов. Также LLM должна самостоятельно принять решение о том, как действовать в каждом из этих случаев, и выполнить эти действия
- Feature Engineering. Агент должен предложить и создать минимум 2 новых признака на основе существующих (например, отношение двух числовых колонок, бинаризация, извлечение дня недели из даты и другие подходы)
- Выбор и обучение модели. На основе ваших данных агент должен предложить с обоснованием несколько наиболее релевантных моделей, выполнить подготовку данных к применению ML-моделей (разделение на признаки и таргет, разделение на train / val / test) и далее обучить эти модели на тренировочной выборке
- Оценка качества. Агент должен вычислить метрики на тестовой выборке
- Отправка результата. Агент должен сформировать отчет с метриками и предсказаниями, а также выводами по моделям. **Примечание**: можно формировать данные в виде структурированного файла, можно создать веб-интерфейс и вывести на нем информацию и заключение о работе моделей

Наконец, реализуйте "память" агента. Ваш ИИ-агент должен сохранять контекст действий в переменных workflow в рамках одного запуска. Например:
- Кратковременная память — агент помнит, какие модели он уже обучил в рамках работы с датасетом в текущей сессии
- Долговременная память — агент сохраняет лучшую модель (например, выгрузка в виде .pkl файла) и метрики в отдельный файл. При следующем запуске агент сможет сравнить новую модель с сохраненной лучшей

В качестве результата выполнения группового проекта ваша команда должна получить презентацию, а также файлы с архитектурой агента и итогами его работы. Демонстрация решения и защита презентации перед комиссией является обязательной и осуществляется в фиксированную дату.

Например, вставлять в презентацию строчки кода или злоупотреблять скринами с работой воркфлоу на различных платформах не рекомендуется.

С точки зрения концепции выполнения проекта — вам необходимо принять на себя роль ML-инженера, который проектирует и внедряет автономные системы машинного обучения. Но помните, что ваша цель — не просто обучить модель, а создать систему, которая автоматизирует процесс принятия решений о том, какую ML-модель, на каких данных и как обучать

Разумеется, чем сложнее и чем нетривиальнее будет ваша исходная постановка задачи, а также чем глубже и чем осмысленнее будут выводы, извлеченные в рамках этой задачи как с помощью агента, так и по результатам вашего анализа, — тем выше окажется качество вашего проекта и тем выше он будет оценен!

* Данный этап является обязательным, что означает, что без его полного выполнения ***нельзя получить оценку за проект выше 1***

#### Артефакты по окончании работы над проектом

* Опционально: веб-страница с агрегированными результатами работы ИИ-агента

## Критерии оценки проекта

Оценка за групповой проект в целом — то есть первая из двух оценок — выставляется по 10-балльной шкале, исходя из 3 критериев. Также комиссия на своё усмотрение может добавить (но не снизить) некоторое количество бонусных баллов за рамками данных критериев.

Каждый из критериев будет перечислен и рассмотрен ниже.

### Критерий 1. Выбор задачи, первичная обработка и качество ML-решения (до 2,5 баллов)

Данный критерий оценивает, насколько обоснованной и осмысленной получилась решаемая бизнес-задача, релевантные ли данные были подобраны для ее решения, насколько полно был реализован весь пайплайн подготовки данных и обучения моделей.

*По данному критерию можно получить **больше 1 балла** только при условии, что итоговый датасет соответствует описанным в обязательных требованиях условиям (см. "Обязательные требования к структуре и объему данных")*

*По данному критерию можно получить **1,5 балла и выше** только при условии, что агент самостоятельно (на основе LLM) провел первичный анализ данных, подготовил их к решению ML-задачи и обучил модели*

*По данному критерию можно получить **2 балла и выше** только при условии, что было выполнено сравнение ML-моделей и были проинтерпретированы результаты решения ML-задачи, полезные с точки зрения бизнеса*

*По данному критерию можно получить **2,5 балла** только при условии, что ИИ-агент использует дополнительные инструменты за рамками базового пайплайна обучения модели (например, подбор оптимальных гиперпараметров ML-моделей)*

### Критерий 2. Архитектура и реализация ИИ-агента (до 4 баллов)

Данный критерий оценивает техническую реализацию ИИ-агента, а именно насколько созданная система соответствует определению ИИ-агента (автономность, использование инструментов, наличие памяти), а также сложность и логичность построенной архитектуры (линейный поток, параллельные ветвления, циклы, обработка ошибок)

***Важно**: в случае оценки **1 балл и ниже** за этот критерий оценка за весь проект становится равна оценке за этот критерий*

*По данному критерию можно получить **больше 1 балла** только при условии соответствия всем обязательным техническим критериям по обязательному использованию инструментов для создания ИИ-агентов (см. требование "Обязательное использование инструментов для создания ИИ-агентов")*

*По данному критерию можно получить **2 балла и выше** только при условии, что в архитектуре агента реализована кратковременная память (агент помнит предыдущие шаги в рамках текущей сессии) И агент использует минимум 2 внешних инструмента (например, доступ к файловой системе, вызов sklearn, отправку уведомлений)*

*По данному критерию можно получить **4 балла** только при условии, что дополнительно реализована долговременная память (сохранение лучшей модели и метрик между запусками, сравнение с предыдущими результатами) И архитектура содержит нелинейные элементы (условные ветвления, циклы, параллельное выполнение задач) и обработку ошибок*

### Критерий 3. Разнообразие использования LLM и техник промптинга (до 3,5 баллов)

Данный критерий оценивает глубину и разнообразие использованных в процессе принятия решений агентом техник для улучшения работы LLM-моделей. В рамках этого критерия рассматривается использование различных техник промптинга, эксперименты с различными LLM, настройка параметров LLM-моделей

*По данному критерию можно получить **больше 1 балла** только при условии использования в проекте хотя бы 2 продвинутых техник промптинга, влияющих на решение агента, И эксперимента с как минимум 2 различными LLM-моделями*

*- Настройка не менее 3 ключевых гиперпараметров LLM с обоснованием выбора конкретных значений для баланса между креативностью и шаблонностью ответов агента*
